```python
#load pakages
import numpy as np
import pandas as pd

# import plotting modules
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
# load data from real data from Self-measured data
df=pd.read_csv("Acc.csv",encoding='GBK')
df2=pd.read_csv("Acc2.csv",encoding='GBK')
```

```python
# calculate absolute value and remove gravity constant 
a1 = np.sqrt(df['X']**2+df['Y']**2+df['Z']**2)
a1 = a1-9.81

a2 = np.sqrt(df2['X']**2+df2['Y']**2+df2['Z']**2)
a2 = a2-9.81
```
